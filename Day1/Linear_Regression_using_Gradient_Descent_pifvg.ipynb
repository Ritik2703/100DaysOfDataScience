```python
"""
Gradient Descent Algorithm
Iterative optimization algorithm to find out minima of a function ( Loss Function)

1) Starts with a random m and c values
2) We need to know what is the gradient of the loss w.r.t m and c to understand do we need 
to reduce the value of m/c to reach minima or we need to increase the value of m/c to reach minima
3) Partial derivative of loss w.r.t m and c is done
4) Update the value of m and c using the following equation -
   m-new = m-old + alpha*(- derivative w.r.t m)
   c-new = c-old + alpha*(- derivative w.r.t c)
5) But this suffers from the problem that the change in value of m and c may be drastic. We
   do not want to make drastic change. So we multiply the derivatives with learning rate(alpha).
   learning rate = 0.01,0.001,0.05 ---- 0< lr <1
6) This process will keep on happening till the time basicall loss is minimized
   gradient or slope of the point is zero. 
"""
```

```python
data = pd.read_csv("data.csv")
data.columns
```

```python
X = data['X ']
Y = data['Y']
plt.scatter(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
# Initialize the m and c values
m = 0
c = 0
L = 0.0001
epochs = 10000 # number of iterations

n = float(X.shape[0]) # number of datapoints in X

# Perform gradient descent
for i in range(epochs):
    Y_pred = m*X + c
    Der_m = (-2/n)* sum((Y - Y_pred)*X)
    Der_c = (-2/n)* sum((Y - Y_pred))
    m = m + (-1)*L*Der_m
    c = c + (-1)*L*Der_c
    print("m and c")
    print(m,c)
    print("Der m and c")
    print(Der_m,Der_c)
    
print("final m and c")
print(m,c)
```

```python
Y_pred_new = m*X + c
plt.scatter(X,Y)
plt.plot([min(X),max(X)],[min(Y_pred_new),max(Y_pred_new)],color = 'red')
```

```python
### LINEAR REGRESSION ##################
X = data['X ']
Y = data['Y']
plt.scatter(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
### LEAST SQUARE METHOD OF SOLVING REGRESSION
# Derivation = 0 
# finding formula for m and c

X_mean = np.mean(X)
Y_mean = np.mean(Y)

n = X.shape[0] # number of datapoints in X
a = 0
b = 0
for i in range(n):
    a = a + ((X[i] - X_mean)*(Y[i]- Y_mean)) 
    b = b + (X[i] - X_mean)**2
m = a/b
c = Y_mean - m*X_mean
    
```
